
**Introducing UFuncs**

For many types of operations, NumPy provides a convenient interface into just this kind of statically typed, compiled routine. This is known as a vectorized operation. This can be accomplished by simply performing an operation on the array, which will then be applied to each element. This vectorized approach is designed to push the loop into the compiled layer that underlies NumPy, leading to much faster execution.


```

#Consider following loop based implementation to ind reciprocals for each element of an array
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
```

```


#Same operatoin using UFuncs applying '/' over array elements
%timeit (1.0 / big_array)

```



Vectorized operations in NumPy are implemented via ufuncs, whose main purpose is to quickly execute repeated operations on values in NumPy arrays. UFuncs are extremely flexible – before we saw an operation between a scalar and an array, but we can also operate between two arrays as well as multidimentional arrays.

Computations using vectorization through ufuncs are nearly always more efficient than their counterpart implemented using Python loops, especially as the arrays grow in size. Any time you see such a loop in a Python script, you should consider whether it can be replaced with a vectorized expression.

**Exploring NumPy's UFuncs**

Ufuncs exist in two flavors: unary ufuncs, which operate on a single input, and binary ufuncs, which operate on two inputs.


**1.What are UFuncs in numpy?**

UFuncs (Universal Functions) in NumPy are functions that operate on ndarray objects element-by-element. They are a core feature of NumPy that enable vectorized operations, which are significantly faster and more efficient than traditional Python loops, especially for large datasets.

Key characteristics and uses of UFuncs include:

Vectorization: They allow operations to be applied to entire arrays rather than individual elements, pushing the computation down to optimized C code.
Performance: This vectorized approach leads to much faster execution times compared to explicit Python loops.
Flexibility: UFuncs can handle operations involving a scalar and an array, two arrays, or multi-dimensional arrays.
Types: They come in two forms: unary ufuncs (operating on a single input, like np.absolute()) and binary ufuncs (operating on two inputs, like np.add() or np.multiply()).
Standard Operations: Common arithmetic operations (+, -, *, /, //, **, %) are implemented as ufuncs, allowing them to work intuitively on NumPy arrays.
Specialized Functions: NumPy provides many specialized ufuncs, and even more are available in modules like scipy.special for complex mathematical operations (e.g., gamma functions, error functions).

**2.Which are various attributes of numy arrays object?**

NumPy array objects have several important attributes that describe their characteristics. Based on the notebook, these attributes include:

ndim: This attribute returns the number of dimensions (axes) of the array. For example, a one-dimensional array has ndim=1, a two-dimensional array has ndim=2, and so on.

Example: For a three-dimensional array x3, x3.ndim would output 3.
shape: This returns a tuple indicating the size of each dimension of the array. **For** instance, a 3x4 array would have shape=(3, 4).

Example: For x3, x3.shape would output (3, 4, 5).
size: This attribute provides the total number of elements in the array, which is the product of the elements of the shape tuple.

Example: For x3, x3.size would output 60 (3 * 4 * 5).
dtype: This describes the data type of the elements in the array. Since NumPy arrays are constrained to contain elements of the same type, dtype indicates this uniform type.

Example: For x3 which contains integers, x3.dtype would output int64.
itemsize: This attribute shows the size, in bytes, of each element in the array.

Example: For x3 with int64 elements, x3.itemsize would output 8 bytes (since int64 is 8 bytes).
nbytes: This gives the total size, in bytes, consumed by all elements in the array. It's equivalent to size * itemsize.

Example: For x3, x3.nbytes would output 480 bytes (60 elements * 8 bytes/element).
These attributes are crucial for understanding and manipulating NumPy arrays effectively.
